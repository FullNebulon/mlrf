```python
import pickle
from matplotlib import pyplot as plt
```

```python
def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict

def to_plt(data):
    plt_data = []
    for i in range(len(data)):
        img = data[i].reshape(3,32,32).transpose(1,2,0)
        plt_data.append(img)
    return plt_data

def print_img(i, data, labels):
    plt.imshow(data[i])
    plt.title(labels[i])
    plt.show()
```

```python
DATA_PATH = '../data/cifar-10-batches-py/'

batch_1 = unpickle(DATA_PATH + 'data_batch_1')
meta = unpickle(DATA_PATH + 'batches.meta')
test_batch = unpickle(DATA_PATH + 'test_batch')

batch_1_data = batch_1[b'data']
test_batch_data = test_batch[b'data']
batch_1_labels = [meta[b'label_names'][batch_1[b'labels'][i]] for i in range(len(batch_1[b'labels']))]
test_batch_labels = [meta[b'label_names'][test_batch[b'labels'][i]] for i in range(len(test_batch[b'labels']))]

plt_batch_1 = to_plt(batch_1_data)
plt_test = to_plt(test_batch_data)
```

```python
print_img(0, plt_batch_1, batch_1_labels)
print_img(0, plt_test, test_batch_labels)
```
